```python
import numpy as np
import matplotlib.pyplot as plt
```

# Nastavení

```python
# pocet zobrazovaných period
periodes = 3

# pocty clenu fourierovy rady
n_list = [5, 10, 30]
```

# Vygenerování signalu + Aproximace Fourierovou řadou 

```python
# vygenerovani jedne periody signalu
t_period = np.linspace(-np.pi, np.pi, 200)
y_period = t_period

# vygenerovani signalu signalu
offset = np.pi if periodes % 2 == 0 else 0
t = np.linspace(-np.pi * periodes - offset, np.pi * periodes - offset, 200 * periodes)
y = np.tile(np.linspace(-np.pi, np.pi, 200), periodes)

for N in n_list:
    # vypocet koeficientu
    T = 2*np.pi
    f = 1/T
    a0 = (2/T) * np.trapz(y_period, t_period)
    an = np.zeros(N)
    bn = np.zeros(N)
    for n in range(1, N+1):
        an[n-1] = (2/T) * np.trapz(y_period * np.cos(n * 2 * np.pi * f * t_period), t_period)
        bn[n-1] = (2/T) * np.trapz(y_period * np.sin(n * 2 * np.pi * f * t_period), t_period)

    # vypocet aproximace fourierovy rady
    y_approx = a0/2
    for n in range(1, N+1):
        y_approx += an[n-1] * np.cos(n * 2 * np.pi * f * t) + bn[n-1] * np.sin(n * 2 * np.pi * f * t)

    # vykresleni do grafu
    plt.plot(t, y, label="y(t)")
    plt.plot(t, y_approx, label="Approximation (N={})".format(N))
    plt.legend()
    plt.show()

    # vypis koeficientu
    print("a0 = {:.2e}".format(a0))
    for i, a in enumerate(an):
        print("a{:1} = {:.2e}".format(i + 1, a))
    for i, b in enumerate(bn):
        print("b{:1} = {:.2e}".format(i + 1, b))
```

# Závěr
V této úloze jsme se zabývali aproximací periodického signálu Fourierovou řadou. Nejprve jsem vygeneroval signál s definovaný počtem period (3), který jsme chtěli aproximovat, a poté jsem pomocí algoritmu vypočítal koeficienty Fourierovy řady. Postupně jsem pak aproximaci prováděli pro různý počet členů řady (N = 5, 10, 30). 

Z výsledků vidíme, že aproximace pomocí Fourierovy řady byla pro daný signál velmi přesná a již při malém počtu členů řady byla výsledná křivka velmi blízká původnímu signálu. S narůstajícím počtem členů řady se aproximovaný signál stáva přesnějším.
